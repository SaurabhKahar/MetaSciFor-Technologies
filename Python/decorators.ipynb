```python
def decorator(func):
  def abc():
    print("Hello Saurabh !!!")
    func()
    print("Bye")
  return abc

@decorator
def xyz():
  print("Hey, Hello")

xyz()
```

```python
def log_operation(func):
    def wrapper(a, b):
        result = func(a, b)
        print(f"Adding {a} and {b} gives {result}")
        return result
    return wrapper

# Use the decorator
@log_operation
def add_two_numbers(a, b):
    return a + b

# Call the decorated function
add_two_numbers(5, 3)
```
