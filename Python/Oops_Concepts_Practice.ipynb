```python
#Calculator using Class Object
from re import M
class calculator:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def add(self):
    return self.x + self.y

  def sub(self):
    return self.x - self.y

  def mul(self):
    return self.x * self.y

  def div(self):
    return self.x // self.y

  def mod(self):
    return self.x % self.y

  def power(self):
    return self.x ** self.y

  def sqrt(self):
    return self.x ** 0.5

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
print()

Addition = calculator(x,y)
Subtraction = calculator(x,y)
Multiplication = calculator(x,y)
Division = calculator(x,y)
Modulus = calculator(x,y)
Power = calculator(x,y)
Sqrt = calculator(x,y)

print(f"Addition of {x, y} is: ", Addition.add())
print(f"Subtraction of {x, y} is: ", Subtraction.sub())
print(f"Multiplication of {x, y} is: ", Multiplication.mul())
print(f"Division of {x, y} is: ", Division.div())
print(f"Modulus of {x, y} is: ", Modulus.mod())
print(f"Power of {x, y} is: ", Power.power())
print(f"Square root of {x, y} is: ", Sqrt.sqrt())
```

```python
#Quiz using Class Object
class Question:
    def __init__(self, question, answer):
        self.question = question
        self.answer = answer

questions = [
    Question("Q1. What is 2 + 2? \na) 1\nb) 2 \nc) 3 \nd) 4 \n", "d"),
    Question("Q2. What is 2 * 2? \na) 4\nb) 2 \nc) 3 \nd) 1 \n", "a")
]

score = 0

for q in questions:
    user_input = input(q.question + "Enter answer here:")

    # Check if the user's answer is correct
    if user_input.strip().lower() == q.answer:
        print("Correct\n")
        score += 1
    else:
        print("Incorrect\n")

print(f"Your total score is: {score} out of {len(questions)}")
```
