Q1. Create a student management system project using the classes and objects concept in python which should perform the following operations

1.Accept Student details

2.Display Student Details

3.Search Details of a Student

4.Delete Details of Student

5.Update Student Details

6.Exit

```python
class Student:
    def __init__(self, name, rollno, age, className):
        self.name = name
        self.rollno = rollno
        self.age = age
        self.className = className

class StudentManagementSystem:
    def __init__(self):
        self.students = []

    def acceptStudentDetails(self, name, rollno, age, className):
        student = Student(name, rollno, age, className)
        self.students.append(student)
        print()

    def displayStudentDetails(self):
        if not self.students:
            print("No student details available.")
            return
        for student in self.students:
            print(f"Name: {student.name}")
            print(f"Rollno: {student.rollno}")
            print(f"Age: {student.age}")
            print(f"Class Name: {student.className}")
            print()

    def searchStudentDetails(self, rollno):
        for student in self.students:
            if student.rollno == rollno:
                print(f"Name: {student.name}")
                print(f"Rollno: {student.rollno}")
                print(f"Age: {student.age}")
                print(f"Class Name: {student.className}")
                print()
                return
        print("Student not found")

    def deleteStudentDetails(self, rollno):
        for student in self.students:
            if student.rollno == rollno:
                self.students.remove(student)
                print("Student details deleted successfully")
                print()
                return
        print("Student not found")
        print()

    def updateStudentDetails(self):
        rollno = int(input("Enter rollno of the student to update: "))
        for student in self.students:
            if student.rollno == rollno:
                student.name = input("Enter new name: ")
                student.age = int(input("Enter new age: "))
                student.className = input("Enter new class name: ")
                print("Student details updated successfully")
                print()
                return
        print("Student not found")
        print()

def main():
    system = StudentManagementSystem()

    while True:
        print("1. Accept Student Details")
        print("2. Display Student Details")
        print("3. Search Student Details")
        print("4. Delete Student Details")
        print("5. Update Student Details")
        print("6. Exit")
        print()

        choice = input("Enter choice: ")
        print()

        if choice == "1":
            name = input("Enter name: ")
            rollno = int(input("Enter rollno: "))
            age = int(input("Enter age: "))
            className = input("Enter class name: ")
            system.acceptStudentDetails(name, rollno, age, className)


        elif choice == "2":
            system.displayStudentDetails()

        elif choice == "3":
            rollno = int(input("Enter rollno: "))
            system.searchStudentDetails(rollno)

        elif choice == "4":
            rollno = int(input("Enter rollno: "))
            system.deleteStudentDetails(rollno)

        elif choice == "5":
            system.updateStudentDetails()

        elif choice == "6":
            print('Exiting system')
            break

        else:
            print("Wrong input")

if __name__ == "__main__":
    main()
```

Q2. Nidhi is trying to understand how the phone gets updated with
new features with all the old features already available in the new phone.
Write a program to explain this with the help of inheritance and method
overriding. Hint: Create class Nokia1, then create class Nokia2 and
inherit all the methods from Nokia1 to Nokia2. Also, modify the required
methods in Nokia2 and add a few additional methods as well. In the
same way, you can create Nokia3 and Nokia4 as well.

```python
class Nokia1:
    def __init__(self):
        self.features = ["Calling", "SMS", "Radio"]

    def show_features(self):
        print("Features of Nokia1:")
        for feature in self.features:
            print(f"- {feature}")

class Nokia2(Nokia1):
    def __init__(self):
        # Directly initializing the parent class without `super()`
        Nokia1.__init__(self)
        # Adding new features specific to Nokia2
        self.features.append("Music Player")
        self.features.append("Bluetooth")

    def show_features(self):
        print("Features of Nokia2:")
        for feature in self.features:
            print(f"- {feature}")

    def play_music(self):
        print("Playing music with Nokia2")

    def use_bluetooth(self):
        print("Using Bluetooth with Nokia2")

# Demonstration
nokia1 = Nokia1()
nokia1.show_features()

print()

nokia2 = Nokia2()
nokia2.show_features()
nokia2.play_music()
nokia2.use_bluetooth()
```
